```python
import random
menu_for_restaurant={101:['Tandoori Chicken',4,240,10],
                     102:['Vegan Burger',1,320,5],
                     103:['Truffle Cake',500,900,20]}

class FOA:
    def menu(self):
        a = input("Do you want to login or check the menu of the restaurant.... \n Please use 'Login' or 'Check Menu'")
        if(a=='Login'):
            b = input('Are you a Admin or Borrower please reply with Admin or Borrower....')
            if (b=='Admin'):
                print('Please use Admin() class to login...')
                

            elif (b=='Borrower'):
                print('Please use Borrower() class to login....')
                

            else:
                print("Please use correct Stakeholder its either 'Admin' or 'Borrower'")

        elif(a=='Check Menu'):
            for key,value in menu_for_restaurant.items():
                print(key,value)
            
        else:
            print('Please choose correct option...')
```

```python
class Admin:
    dict_of_admin={}
    dict_of_login_admin={}

    def  __init__(self):
        print("Welcome Admin.... Please login or signup...")


    def sign_up(self):     
        self.name=input("Please enter your full name.... ")
        self.email=input("Please enter your email address.... ")
        self.password=input("Please enter your password.... ")
        
        if self.name not in Admin.dict_of_admin:
            self.dict_of_admin[self.name]=[self.email,self.password]
            print('Account created sucessfully')

        else:
            print('Name already exist..')
            
            
    def login(self):
        self.name=input("Please enter your name Admin...")
        email=input("Please enter your email address.... ")
        password=input("Please enter your password.... ")
                
        if self.name in Admin.dict_of_admin:
            if(self.email==email and self.password==password):
                Admin.dict_of_login_admin[self.name]=[self.email,self.password]
                print("Login successful")
            else:
                print("Incorrect login email and password")
        else:
            print("You don't have any account ,please sign-up")
            
    def logout(self,adname):
        self.adname=adname
        if self.adname in Admin.dict_of_login_admin:
            print("you are logged out")
            Admin.dict_of_login_admin.pop(self.name)
        else:
            print("you are not logged in")
            
            
    def add_item(self):
        self.admin_name=input("Please enter your name Admin...")
        if self.admin_name in Admin.dict_of_login_admin:
            self.food_name=input('Enter the name of food...')
            self.quantity=int(input('Enter the quantity of food...'))
            self.unit=input('Enter the unit of the food product...')
            self.price=int(input('Enter the price per quantity...'))
            self.available=int(input('Enter the quantity which is available in the restaurant...'))
            self.food_id=random.randint(0,5000)
            menu_for_restaurant[self.food_id]=[self.food_name,self.quantity,self.unit,self.price,self.available]
            print('Food item got added with food ID..',self.food_id)
            
        else:
            print('Please login as an Admin first...')
            
            
            
    def edit_item(self):
        self.admin_name=input("Please enter your name Admin...")
        if self.admin_name in Admin.dict_of_login_admin:
            self.food_id=int(input('Enter ID of the food which you want to edit...'))
            if self.food_id in menu_for_restaurant:
                self.food_name=input('Enter the name of food...')
                self.quantity=int(input('Enter the quantity of food...'))
                self.unit=input('Enter the unit of the food product...')
                self.price=int(input('Enter the price per quantity...'))
                self.available=int(input('Enter the quantity which is available in the restaurant...'))
                menu_for_restaurant[self.food_id]=[self.food_name,self.quantity,self.unit,self.price,self.available]
                print('Food item edited sucessfully heving food ID..',self.food_id)

            else:
                print('Item with food ID {} does not exits...'.format(self.food_id))
                
        else:
            print('Please first login as an Admin...')
            
                       
            
    def delete_item(self):
        self.admin_name=input("Please enter your name Admin...")
        if self.admin_name in Admin.dict_of_login_admin:
            self.food_id=int(input('Enter ID of the food which you want to delete...'))
            if self.food_id in menu_for_restaurant:
                menu_for_restaurant.pop(self.food_id)
                print('Food item deleted successfully....')
            
            else:
                print('Please choose correct food ID...')

        else:
            print('Please first login as an Admin...')     
            
            
            
    def logout(self,):
        self.adname=input('Enter your name admin...')
        if self.adname in Admin.dict_of_login_admin:
            print("you are logged out")
            Admin.dict_of_login_admin.pop(self.name)
        else:
            print("you are not logged in")
```

```python
class User:
    user_details={}
    login_users={}
    order_history={}
    
    def __init__(self):
        print('Hello!! if you are a new user then register or else login')
        
    def register(self):
        self.name=input('Enter your full name here...')
        self.phone_no=input('Enter your phone no...')
        self.email=input('Enter your email address...')
        self.address=input('Enter your address...')
        self.password=input('Enter you password...')
        
        if self.email not in User.user_details:
            User.user_details[self.email]=[self.password,self.name,self.phone_no, self.address]
            print('Registered successfully...')


        else:
            print('Email already taken...')


    def login(self):
        self.email=input('Enter your email address...')
        if self.email in User.user_details:
            self.password=input('Enter your password...')
            if self.password==User.user_details[self.email][0]:
                User.login_users[self.email]=self.password
                print('Login successfullly...')
                print('You further can proceed for :\n 1. Placing an order \n 2. View order history \n 3. Update your profile')
                    
            
            else:
                print('Incorrect password...')
                
        else:
            print('This email does not exists please sign up first...')
            
            
    def update_profile(self):
        self.email=input('Enter your email address...')
        if self.email in User.user_details:
            self.password=input('Enter your password...')
            if self.password==User.user_details[self.email][0]:
                self.name=input('Enter your full name here...')
                self.phone_no=input('Enter your phone no...')
                self.address=input('Enter your address...')
                self.new_password=input('Enter your new password...')
                User.user_details[self.email]=[self.new_password,self.name,self.phone_no, self.address]
                print('Data updated successfully...')


            else:
                print('Incorrect password...')
                
        else:
            print('This email does not exist...')
            
            
    def food_list(self):
        for key,value in menu_for_restaurant.items():
            print(value[0],"({} {}) [INR {}]".format(value[1],value[2],value[3]))
            
            
    def place_order(self):
        self.email=input('Enter your email address...')
        if self.email in User.login_users:
            self.food_id=input('Enter food ID of all the food items that you want to purchase...').split(' ')
            self.food_id=list(map(int,self.food_id))
            self.total=0
            for i in self.food_id:
                if i in menu_for_restaurant:
                    print('Enter quantity of {} in {}'.format(menu_for_restaurant[i][0],menu_for_restaurant[i][2]))
                    self.qua=int(input('Enter Quantity'))
                    if self.qua < menu_for_restaurant[i][4]:
                        self.amt=(menu_for_restaurant[i][3]/menu_for_restaurant[i][1])*self.qua
                        print('Amount for this dish is RS.',self.amt)
                        menu_for_restaurant[i][4]-=self.qua
                        self.total+=self.amt            
                    else:
                        print('This dish is currently not available')

                else:
                    print('Incorrect food id..',i)

            print('Your bill amount is RS.',self.total)

            self.order=input('To place order write Yes ')
            if self.order=='Yes' or self.order =='yes':
                print('Order placed succesfully... \n Your bill Amt is RS.',self.total)
                print('Thanks for ordering...')
                
                if self.email in User.order_history:
                    User.order_history[self.email].append([self.food_id,self.total])
                   
                    
                else:
                    User.order_history[self.email]=[self.food_id,self.total]
                    
            else:
                print('Please write yes to place order...')
                    
        else:
            print('Email does not exist...')
                 
            

                
    def history_of_order(self):
        self.email=input('Enter your email address...')
        if self.email in User.login_users:
            if self.email in User.order_history:
                print("Your Order History: ",User.order_history[self.email])
                    
                
            else:
                print("You have not placed any order...")
                
        else:
            print('Please login to see your order history...')


            
            
```
